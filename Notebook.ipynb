### 3.1
This section is about variables and basic operations.

Answer:
Variables are used to store data in Python. You assign values using the = operator.

```python
x = 10
y = 5
print(x + y)
```

### 3.2
This section is about strings.

Answer:
Strings are sequences of characters in Python and are written using quotes.

```python
name = "Best"
print("Hello", name)
```

### 3.3
This section is about lists.

Answer:
Lists are used to store multiple items in a single variable.

```python
numbers = [1, 2, 3, 4]
print(numbers)
```

### 3.4
This section is about conditionals.

Answer:
Conditionals allow programs to make decisions using if statements.

```python
x = 10
if x > 5:
    print("x is greater than 5")
```

### 3.5
This section is about loops.

Answer:
Loops allow repeating actions multiple times.

```python
for i in range(5):
    print(i)
```

### 3.6
This section is about functions.

Answer:
Functions are reusable blocks of code that perform a specific task.

```python
def greet(name):
    print("Hello", name)

greet("Best")
```
