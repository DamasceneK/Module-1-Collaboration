# Sort an Array of 0s, 1s, and 2s

This problem was solved using a class `Solution` with a method `sort012` that:
- Counts the number of 0s, 1s, and 2s
- Overwrites the original array in sorted order
- Returns nothing (modifies in place)

```python
class Solution:
    def sort012(self, arr):
        count0 = arr.count(0)
        count1 = arr.count(1)
        count2 = arr.count(2)

        i = 0

        for _ in range(count0):
            arr[i] = 0
            i += 1

        for _ in range(count1):
            arr[i] = 1
            i += 1

        for _ in range(count2):
            arr[i] = 2
            i += 1


# Example test
arr = [0, 1, 2, 0, 1, 2]
Solution().sort012(arr)
print(arr)
```

## Exercise 2: Binary Search
- Used a function to implement binary search
- Returns index if found, else -1

```python
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid      # index of target
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1                 # not found


# Test the function
numbers = [1, 3, 5, 7, 9, 11, 13]
target = 7

result = binary_search(numbers, target)
if result != -1:
    print(f"Element found at index {result}")
else:
    print("Element not found")
```
