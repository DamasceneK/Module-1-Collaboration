4.1 Choose a number between 1 and 10

```python
# Assign values
secret = 7
guess = 5

# Conditional checks
if guess < secret:
    print("too low")
elif guess > secret:
    print("too high")
else:
    print("just right")
```

4.2 Assign True or False to the variables small and green

```python
# Assign True or False
small = True
green = True

# Check combinations
if small and green:
    print("pea")
elif small and not green:
    print("cherry")
elif not small and green:
    print("watermelon")
else:
    print("pumpkin")
```

```python
# List of values
numbers = [3, 2, 1, 0]

# For loop to print each value
for num in numbers:
    print(num)
```

```python
6.2 Assign the value 7 to the variable guess_me, and the value 1 to the variable numbe
```

```python
# Assign values
guess_me = 7
number = 1

# While loop
while True:
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break  # exit loop
    else:  # number > guess_me
        print("oops")
        break  # exit loop
    number += 1  # increment number
```

```python
# Assign value
guess_me = 5

# For loop over range(10)
for number in range(10):
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break  # exit loop
    else:  # number > guess_me
        print("oops")
        break  # exit loop
```
