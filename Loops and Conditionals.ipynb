4.1 Choose a number between 1 and 10

```python
# Assign values
secret = 7
guess = 5

# Conditional checks
if guess < secret:
    print("too low")
elif guess > secret:
    print("too high")
else:
    print("just right")
```

4.2 Assign True or False to the variables small and green

```python
# Assign True or False
small = True
green = True

# Check combinations
if small and green:
    print("pea")
elif small and not green:
    print("cherry")
elif not small and green:
    print("watermelon")
else:
    print("pumpkin")
```
