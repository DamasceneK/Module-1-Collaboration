```python
things = ["mozzarella", "cinderella", "salmonella"]
print(things)
```

 7.5

The element that refers to a person is **"cinderella"**.  
After capitalizing it and printing the list, yes, the element in the list changes permanently because lists are mutable.

```python
things[1] = things[1].capitalize()
print(things)
```

 7.6

The cheesy element is **"mozzarella"**.  
Changing it to uppercase updates the list element directly.

```python
things[0] = things[0].upper()
print(things)
```
